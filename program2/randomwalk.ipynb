```python
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd

from string import digits
```

```python
data=pd.read_excel('following.xlsx')
data.head()
```

```python
def getname(x):
    x=x.replace('https://www.researchgate.net/profile/','')
#     remove_digits = str.maketrans('', '', digits)
#     x=x.translate(remove_digits)
    return x.replace('_',' ')

data['Researcher']=data['ResearcherURL'].apply(lambda x:getname(x))
data['Following']=data['FollowingURL'].apply(lambda x:getname(x))
```

```python
data.to_excel('following_clean.xlsx',index=None)
```

```python
%matplotlib notebook
net_grid = nx.Graph()   
list_net_nodes = all_node
list_net_edges =all_list
net_grid.add_nodes_from(list_net_nodes)
net_grid.add_edges_from(list_net_edges)
nx.draw(net_grid, with_labels=True)
plt.show()
```

```python
%%time
all_node_dict={}
for i in all_node:
    temp={}
    
    data_A=data[data.Researcher==i]
    
    data_B=data[data.Following==i]

    if data_A.shape[0]!=0:
        for j in list(set(data_A['Following'].unique().tolist())):
            temp[j]=1
    if data_B.shape[0]!=0:
        for k in list(set(data_B['Researcher'].unique().tolist())):
            temp[k]=1
    all_node_dict[i]=temp
```

```python

def PersonalRank(G, max_sim, root, max_step):
    alpha=0.85
    rank = dict()
    rank = {x:0 for x in G.keys()}
    rank[root] = 1
    for k in range(max_step):
        tmp = {x:0 for x in G.keys()}
        for i, ri in G.items():
                tmp[j] += alpha * rank[i] / (1.0 * len(ri))
        tmp[root] += (1 - alpha)
        rank = tmp
    rankvalue=rank.values()
    rankvalue_max=max(rankvalue)
    rankvalue_min=min(rankvalue)
    for key, value in rank.items():
        rank[key]=  (value - rankvalue_min) / (rankvalue_max - rankvalue_min)
    rank=sorted(rank.items(), key=lambda d:d[1], reverse=True)[1:max_sim+1]
 
    return rank
```
